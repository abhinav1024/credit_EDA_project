```python
def plotting(train, train0, train1, column): # Function created for doing Univariate analysis
    
    train = train
    train_0 = train0
    train_1 = train1
    col = column
    
    fig = plt.figure(figsize=(13,10))
    
    ax1 = plt.subplot(221)
    train[col].value_counts().plot.pie(autopct = "%1.0f%%", ax=ax1)
    plt.title('Plotting data for the column: '+ column,fontdict={'fontsize':15})
    
    ax2 = plt.subplot(222)
    sns.countplot(x= column, hue = 'TARGET', data = train, ax = ax2)
    plt.xticks(rotation=90)
    plt.title('Plotting data for target in terms of total count',fontdict={'fontsize':15})


    ax3 = plt.subplot(223)
    df = pd.DataFrame()
    df['0']= ((train_0[col].value_counts())/len(train_0))
    df['1']= ((train_1[col].value_counts())/len(train_1))
    df.plot.bar(ax=ax3)
    plt.title('Plotting data for target in terms of percentage')


    fig.tight_layout() # Or equivalently,  "plt.tight_layout()"

    plt.show()
      
```

```python
def plot_chart(column, hue):
    col = column
    hue = hue
    fig = plt.figure(figsize=(11,11))
    
    ax1 = plt.subplot(221)
    applications_merged[col].value_counts().plot.pie(autopct="%1.0f%%", ax=ax1)
    plt.title('Plotting data for the column: '+column)
    
    ax2 = plt.subplot(222)
    df = pd.DataFrame()
    df['0'] = ((t0[col].value_counts())/len(t0))
    df['1'] = ((t1[col].value_counts())/len(t1))
    df.plot.bar(ax=ax2)
    plt.title('Plotting data for target in terms of total percetage...')
    
    ax3 = plt.subplot(223)
    sns.countplot(x=col, hue=hue, data=t0, ax=ax3)
    plt.xticks(rotation=90)
    plt.title('Plotting data for Target=0 in terms of percentage...')
    
    ax4 = plt.subplot(224)
    sns.countplot(x=col, hue=hue, data=t1, ax=ax4)
    plt.xticks(rotation=90)
    plt.title('Plotting data for Target=1 in terms of percentage...')
    
    fig.tight_layout()
    plt.show()
```

```python
# Analyse NAME_EDUCATION_TYPE vs NAME_CONTRACT_STATUS
plot_chart('NAME_EDUCATION_TYPE','NAME_CONTRACT_STATUS')
```

**Insights:**
1. Secondary/Secondary special are the highest applicants for loan
2. However, Secondary/Secondary special have more difficulty in repaying loan then any other categories.

```python
# Analyse NAME_INCOME_TYPE vs NAME_CONTRACT_STATUS
plot_chart('NAME_INCOME_TYPE','NAME_CONTRACT_STATUS')
```

**Insights:**
1. Although the Working class clients have applied the most(61%) for the loan, they still tend to find it more difficult to re-pay their loans on time. 
2. That is why Working class clients with payment difficulty have higher rejections & cancellations than approvals.
